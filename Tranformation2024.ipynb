```python
%sql
CREATE OR REPLACE VIEW delayed_flights AS
SELECT *
FROM flights
WHERE ARR_DELAY > 0;
```

```python
%sql
SELECT * FROM delayed_flights limit 5;
```

```python
%sql
DROP VIEW IF EXISTS delayed_flights;
```

```python
%sql
CREATE OR REPLACE VIEW ground_delay AS
SELECT *,
       GREATEST(DEP_DELAY, 0) + GREATEST(TAXI_OUT, 0) AS GROUND_DELAY
FROM delayed_flights;
```

```python
%sql
CREATE OR REPLACE VIEW air_delay AS
SELECT *,
       GREATEST(AIR_TIME - CRS_ELAPSED_TIME, 0) AS AIR_DELAY
FROM ground_delay;
```

```python
%sql
CREATE OR REPLACE VIEW arrival_delay AS
SELECT *,
       GREATEST(TAXI_IN, 0) AS ARRIVAL_CONGESTION_DELAY
FROM air_delay;
```

```python
%sql
CREATE OR REPLACE VIEW validated_delays AS
SELECT *,
       GROUND_DELAY + AIR_DELAY + ARRIVAL_CONGESTION_DELAY AS TOTAL_COMPONENT_DELAY,
       ARR_DELAY - (GROUND_DELAY + AIR_DELAY + ARRIVAL_CONGESTION_DELAY) AS UNEXPLAINED_DELAY
FROM arrival_delay;
```

```python
%sql
CREATE OR REPLACE TABLE weather_flagged AS
SELECT *,
       CASE
           WHEN O_PRCP > 0 OR D_PRCP > 0
             OR O_WSPD > 20 OR D_WSPD > 20
           THEN 1
           ELSE 0
       END AS WEATHER_FLAG
FROM validated_delays;
```

```python
%sql
DESCRIBE weather_flagged;
```

```python
%sql
SELECT
    AVG(GROUND_DELAY) AS AVG_GROUND_DELAY,
    AVG(AIR_DELAY) AS AVG_AIR_DELAY,
    AVG(ARRIVAL_CONGESTION_DELAY) AS AVG_ARRIVAL_CONGESTION
FROM weather_flagged;
```

```python
%sql
SELECT
    WEATHER_FLAG,
    AVG(ARR_DELAY) AS AVG_ARR_DELAY
FROM weather_flagged
GROUP BY WEATHER_FLAG;
```

```python
%sql
SELECT
    OP_CARRIER,
    AVG(GROUND_DELAY) AS AVG_GROUND_DELAY,
    AVG(AIR_DELAY) AS AVG_AIR_DELAY,
    AVG(ARRIVAL_CONGESTION_DELAY) AS AVG_ARRIVAL_DELAY
FROM weather_flagged
GROUP BY OP_CARRIER
ORDER BY AVG_GROUND_DELAY DESC;
```

```python
%sql
SELECT
    DEST,
    AVG(ARRIVAL_CONGESTION_DELAY) AS AVG_ARRIVAL_CONGESTION
FROM weather_flagged
GROUP BY DEST
ORDER BY AVG_ARRIVAL_CONGESTION DESC;
```

```python
%sql
SELECT
    AVG(GROUND_DELAY) AS GROUND_DELAY,
    AVG(AIR_DELAY) AS AIR_DELAY,
    AVG(ARRIVAL_CONGESTION_DELAY) AS ARRIVAL_CONGESTION_DELAY
FROM validated_delays;
```
